```python
result=[]
col=['GAME_ID','championId','lane','teamId','kda','killParticipation','totalDamageDealtToChampions']
for i in range(len(df)):
    lst=[]
    lst.append(df.iloc[i]['GAME_ID'])
    lst.append(df.iloc[i]['MATCHES']['champion']['championId'])
    lst.append(df.iloc[i]['MATCHES']['champion']['teamPosition'])
    lst.append(df.iloc[i]['MATCHES']['champion']['teamId'])
    lst.append(df.iloc[i]['MATCHES']['kda']['kda'])
    lst.append(df.iloc[i]['MATCHES']['kda']['killParticipation'])
    lst.append(df.iloc[i]['MATCHES']['damage']['totalDamageDealtToChampions'])
    
    result.append(lst)
match_info = pd.DataFrame(result,columns=col)
    
```
