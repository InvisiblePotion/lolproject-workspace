```python
result=[]
col=['championId','lane','runetype_core','core_mainrune','core_rune1','core_rune2','core_rune3','runetype_sub','sub_rune1','sub_rune2','shard_offense','shard_flex','shard_defence','win']
for i in range(len(df)):
    lst=[]
    lst.append(df.iloc[i]['MATCHES']['champion']['championId'])
    lst.append(df.iloc[i]['MATCHES']['champion']['teamPosition'])
    lst.append(df.iloc[i]['MATCHES']['rune']['runePrimaryStyle'])
    lst.append(df.iloc[i]['MATCHES']['rune']['runeCorePerk'])
    lst.append(df.iloc[i]['MATCHES']['rune']['runePrimaryPerk1'])
    lst.append(df.iloc[i]['MATCHES']['rune']['runePrimaryPerk2'])
    lst.append(df.iloc[i]['MATCHES']['rune']['runePrimaryPerk3'])
    lst.append(df.iloc[i]['MATCHES']['rune']['runeSubStyle'])
    lst.append(df.iloc[i]['MATCHES']['rune']['runeSubPerk1'])
    lst.append(df.iloc[i]['MATCHES']['rune']['runeSubPerk2'])
    lst.append(df.iloc[i]['MATCHES']['rune']['runeShardOffense'])
    lst.append(df.iloc[i]['MATCHES']['rune']['runeShardFlex'])
    lst.append(df.iloc[i]['MATCHES']['rune']['runeShardDefense'])
    lst.append(df.iloc[i]['MATCHES']['champion']['win'])
    
    result.append(lst)
    champ_rune = pd.DataFrame(result,columns=col)
```

```python
champ_rune
```

```python
champ_rune_count = champ_rune.groupby(['championId','lane','runetype_core','core_mainrune','runetype_sub']).count().rename(columns={'win':'games'})[['games']]
```

```python
champ_rune_win = champ_rune.groupby(['championId','lane','runetype_core','core_mainrune','runetype_sub']).sum()[['win']]
```

```python
result = champ_rune_count.join(champ_rune_win)
result
```

```python
result['win_rate']= round(result.win/result.games * 100,2)
```
