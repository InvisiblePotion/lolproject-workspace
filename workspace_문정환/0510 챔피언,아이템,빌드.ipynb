```python
now = datetime.datetime.now()
now = now.strftime('%Y-%m-%d %H:%M:%S')
```

```python
df = mu.oracle_totalExecute('select version,game_id,champion_id,participant_number,champion,item,itemtree from rawdata')
df = df.sort_values(['GAME_ID','PARTICIPANT_NUMBER'])
df_cols = df.columns[4:]

matches = {a:[] for a in df_cols}
for rec in tqdm(df.iloc):
    for col in [a for a in df_cols]:
        matches[col].append(rawdataLoader(rec[col]))
for col in [a for a in df_cols]:
    df[col] = matches[col]


version_lst = df['VERSION'].to_list()
champid_lst = df['CHAMPION_ID'].to_list()
champ_lst = df['CHAMPION'].to_list()
item_lst = df['ITEM'].to_list()
itemtree_lst = df['ITEMTREE'].to_list()

result=[]
col=['version','update_time','champ_id','lane','win','item0','item1','item2','item3','item4','item5','item6','core_item']
for idx,champ in enumerate(tqdm(champ_lst)):  
    result.append([
        version_lst[idx],
        now,
        champid_lst[idx],
        champ['teamPosition'],
        champ['win'],
        item_lst[idx]['item0'],
        item_lst[idx]['item1'],
        item_lst[idx]['item2'],
        item_lst[idx]['item3'],
        item_lst[idx]['item4'],
        item_lst[idx]['item5'],
        item_lst[idx]['item6'],
        str(itemtree_lst[idx])
    ])
    
champ_item_build= pd.DataFrame(result,columns=col)
```

```python
df = mu.oracle_totalExecute('select version,game_id,champion_id,participant_number,champion,itemtree from rawdata')
df = df.sort_values(['GAME_ID','PARTICIPANT_NUMBER'])
df_cols = df.columns[4:]

matches = {a:[] for a in df_cols}
for rec in tqdm(df.iloc):
    for col in [a for a in df_cols]:
        matches[col].append(rawdataLoader(rec[col]))
for col in [a for a in df_cols]:
    df[col] = matches[col]


version_lst = df['VERSION'].to_list()
champid_lst = df['CHAMPION_ID'].to_list()
champ_lst = df['CHAMPION'].to_list()
itemtree_lst = df['ITEMTREE'].to_list()

result=[]
col=['version','update_time','champ_id','lane','win','item1','item2','item3']
for i in tqdm(range(len(df))):
    lst=[]
    lst.append(version_lst[i])
    lst.append(now)
    lst.append(champid_lst[i])
    lst.append(champ_lst[i]['teamPosition'])
    lst.append(int(champ_lst[i]['win']))
    for j in range(len(itemtree_lst[i])):
        lst.append(int(itemtree_lst[i][j]))
    
    result.append(lst)

champ_item_build= pd.DataFrame(result,columns=col)
champ_item_build.fillna(0,inplace=True)
champ_item_build = champ_item_build.astype({'item1':int,'item2':int,'item3':int})
```

```python
champ_item_build_count = champ_item_build.groupby(['version','update_time','champ_id','lane','item1','item2','item3']).count()[['win']].rename(columns={'win':'game_count'})#.sort_values('game_count',ascending=False)
```

```python
champ_item_build_win = champ_item_build.groupby(['version','update_time','champ_id','lane','item1','item2','item3']).sum()[['win']]#.sort_values('win',ascending=False)
```

```python
result = champ_item_build_count.join(champ_item_build_win)#.sort_values('game_count',ascending=False)
```

```python
result.reset_index(inplace=True)
```
