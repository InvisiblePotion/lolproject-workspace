```python
# import team_utils as tu
import pandas as pd
import numpy as np
import requests
import my_utils2 as mu
from tqdm import tqdm
import json
import datetime
```

```python
def rawdataLoader(datas: str):
    return json.loads(datas.replace("'", '"').replace('True', '1').replace('False', '0'))
```

```python
now = datetime.datetime.now()
now = now.strftime('%Y-%m-%d %H:%M:%S')
```

```python
df = mu.oracle_totalExecute("SELECT version,game_id, participant_number,champion_id, champion ,item FROM RAWDATA where version = '13.9.506.4846'")
    
df = df.sort_values(['GAME_ID','PARTICIPANT_NUMBER'])
df_cols = df.columns[4:]

matches = {a:[] for a in df_cols}
for rec in tqdm(df.iloc):
    for col in [a for a in df_cols]:
         matches[col].append(rawdataLoader(rec[col]))
for col in [a for a in df_cols]:
    df[col] = matches[col]


#list화 속도개선  
version_list = df['VERSION'].tolist()
gameid_list = df['GAME_ID'].tolist()
championid_lst = df['CHAMPION_ID'].tolist()
champion_list =  df['CHAMPION'].tolist()
item_list = df['ITEM'].tolist()

result = []
col=['version','update_time','championId','lane','boots','win']
for idx,champ in enumerate(tqdm(champion_list)):
    result.append([
        version_list[idx],
        now,
        championid_lst[idx],
        champ['teamPosition'],
        list(item_list[idx].values()),
        int(champ['win'])
    ])
champ_boots = pd.DataFrame(result,columns=col)
```

```python
boots_list = [1001, 2422, 3006, 3009, 3020, 3047, 3111, 3117, 3158]
new_boots = []
for x in champ_boots['boots'].values:
    if x in boots_list:
        new_boots.append(x)
    else:
        new_boots.append(0)
champ_boots['boots1'] = new_boots
```

```python
#신발 뽑아내기
champ_boots['boots'] = champ_boots['boots'].apply(lambda lst: [x for x in lst if x in [1001, 2422, 3006, 3009, 3020, 3047, 3111, 3117, 3158]])
```

```python
#언패킹
champ_boots['boots'] = champ_boots['boots'].apply(lambda lst: lst[0] if len(lst) > 0 else 0)
```

```python
champ_boots.drop('boots1',axis=1,inplace=True)
```

```python
champ_boots_count = champ_boots.groupby(['version','update_time','championId','lane','boots']).count().rename(columns={'win':'game_count'})
```

```python
champ_boots_win = champ_boots.groupby(['version','update_time','championId','lane','boots']).sum()
```

```python
champ_boots_win.sort_values('win',ascending=False)
```

```python
result = champ_boots_count.join(champ_boots_win)
```

```python
result.reset_index(inplace=True)
```

```python
result.sort_values('win',ascending=False)
```

```python
result['win_rate']=round(result.win/result.game_count*100.2)
```

```python
result['pick_rate'] = round(result.game_count/(sum(result.game_count)/10)*100,2)
```
