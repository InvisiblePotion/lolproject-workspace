```python
import team_utils as tu
import pandas as pd
import requests
from tqdm import tqdm
```

```python
url = 'http://ddragon.leagueoflegends.com/cdn/13.8.1/data/ko_KR/champion.json'

def ChampSkin(url):
    res = requests.get(url).json()
    df_origine = pd.DataFrame(res)
    
    df_info = []
    for k in range(len(df_origine['data'])):
        lst1=[]
        lst1.append(int(df_origine.iloc[k]['data']['key']))
        lst1.append(df_origine.iloc[k]['data']['id'])
        lst1.append(df_origine.iloc[k]['data']['name'])
        
        df_info.append(lst1)
    col = ['championId','championName','championName_KR']
    df_info = pd.DataFrame(df_info,columns=col)
    
    result=[]
    col = ['champion_id','skin_id','skin_num','skin_name','skin_chromas']
    for i in tqdm(range(len(df_info))):
        lst3=[]
        name = df_info.iloc[i]['championName']
        url2 = f'http://ddragon.leagueoflegends.com/cdn/13.8.1/data/ko_KR/champion/{name}.json'
        res2 = requests.get(url2).json()

        for j in res2['data'][name]['skins']:
            lst2=[]
            lst2.append(df_info.iloc[i]['championId'])
            lst2.append(j['id'])
            lst2.append(j['num'])
            lst2.append(j['name'])
            if j['chromas'] == False:
                lst2.append(0)
            elif j['chromas'] == True:
                lst2.append(1)

            result.append(lst2)
        df=pd.DataFrame(result,columns=col)
    return df
```
