```python
from tqdm import tqdm
import pandas as pd
tqdm.pandas()
import random
import requests
import time
import numpy as np
```

```python
url='http://ddragon.leagueoflegends.com/cdn/13.8.1/data/ko_KR/summoner.json'

def summonerSpell(url):
    
    res= requests.get(url).json() #데이터 불러오기
    spell_df = pd.DataFrame(res) # 데이터프레임화
    
    # 원하는값 뽑아서 다시 데이터프레임화 시키기
    result = []
    col = ['spell','name','key','description','tooltip','cooldownBurn','img_url']
    for i in range(len(spell_df)):
        lst=[]
        lst.append(spell_df.iloc[i]['data']['id'])
        lst.append(spell_df.iloc[i]['data']['name'])
        lst.append(int(spell_df.iloc[i]['data']['key']))
        lst.append(spell_df.iloc[i]['data']['description'])
        lst.append(spell_df.iloc[i]['data']['tooltip'])
        lst.append(spell_df.iloc[i]['data']['cooldownBurn'])
        lst.append('http://ddragon.leagueoflegends.com/cdn/13.8.1/img/spell/'+spell_df.iloc[i]['data']['image']['full'])
    
        result.append(lst)
    spells_df = pd.DataFrame(result,columns=col)  
    
    return spells_df
```

```python
df = summonerSpell(url)
```
