# 아이템 데이터 프레임화

```python
url = 'http://ddragon.leagueoflegends.com/cdn/13.8.1/data/ko_KR/item.json'
def item_table(url):
    
    res = requests.get(url).json()
    
    #아이템 이름 배열에 저장
    item_ls = []
    for i in res['data']:
        item_ls.append(i)
    
    #원하는값 뽑아서 데이터 프레임화
    item_table = pd.DataFrame()
    for j in item_ls:
        item_id = int(j)
        try:
            item_name = res['data'][j]['name']
        except:
            item_name = np.nan

        try:
            item_desc = res['data'][j]['description']
        except:
            item_desc = np.nan

        try:
            item_icon =  'http://ddragon.leagueoflegends.com/cdn/13.8.1/img/item/'+j+'.png'
        except:
            item_icon = np.nan

        try:
            item_plaintext = res['data'][j]['plaintext']
        except:
            item_plaintext = np.nan

        try:
            item_totalgold = res['data'][j]['gold']['total']
        except:
            item_totalgold = np.nan

        try:
            item_basegold = res['data'][j]['gold']['base']
        except:
            item_basegold = np.nan


        pre_df = pd.DataFrame({'item_id' : [item_id],
                               'item_name' : [item_name],
                               'item_desc' : [item_desc],
                               'item_icon' : [item_icon],
                               'item_plaintext' : [item_plaintext],
                               'item_totalgold' : [item_totalgold],
                               'item_basegold' : [item_basegold]
                              })

        item_table = item_table.append(pre_df)
        item_table.reset_index(inplace=True)
        item_table.drop('index',axis=1,inplace=True)
        
    return item_table
```

# upper 아이템 데이터 프레임화

```python
url = 'http://ddragon.leagueoflegends.com/cdn/13.8.1/data/ko_KR/item.json'
def upper_item(url):
    res = requests.get(url).json()
    
    item_ls = []
    for i in res['data']:
        item_ls.append(i)
    
    
    item_upper = pd.DataFrame()
    for j in item_ls:
        
        item_id = int(j)
        
        try:
            next_item = res['data'][j]['into']
        except:
            next_item = np.nan

        upper_df = pd.DataFrame({'item_id' : [item_id],
                               'next_item' : [(next_item)]
                                })

        item_upper = item_upper.append(upper_df)
        item_upper.reset_index(inplace=True)
        item_upper.drop('index',axis=1,inplace=True)
    return item_upper
```
