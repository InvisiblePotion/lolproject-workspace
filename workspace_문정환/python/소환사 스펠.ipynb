# 데이터 불러오기

```python
url='http://ddragon.leagueoflegends.com/cdn/13.8.1/data/ko_KR/summoner.json'
```

```python
res= requests.get(url).json()
res
```

```python
res.keys()
```

```python
spell_df = pd.DataFrame(res)
spell_df
```

```python
spell_df.reset_index(inplace=True)
```

```python
result = []
col = ['spell','name','key','description','tooltip','cooldownBurn','img_url']
for i in range(len(spell_df)):
    lst=[]
    lst.append(spell_df.iloc[i]['data']['id'])
    lst.append(spell_df.iloc[i]['data']['name'])
    lst.append(spell_df.iloc[i]['data']['key'])
    lst.append(spell_df.iloc[i]['data']['description'])
    lst.append(spell_df.iloc[i]['data']['tooltip'])
    lst.append(spell_df.iloc[i]['data']['cooldownBurn'])
    lst.append(spell_df.iloc[i]['data']['image']['full'])
    
    result.append(lst)
    
    spells_df = pd.DataFrame(result,columns=col)    
```
