```python
url='http://ddragon.leagueoflegends.com/cdn/13.8.1/data/ko_KR/runesReforged.json'
def runes(url):
    df = requests.get(url).json()
    
    df_runetype = []
    df_rune = []
    for i in range(len(df)) :
        lst2 = []
        lst2.append(df[i]['id'])
        lst2.append(df[i]['name'])
        lst2.append(df[i]['key'])
        icon2 = str.lower(df[i]['icon'])
        lst2.append('https://raw.communitydragon.org/latest/plugins/rcp-be-lol-game-data/global/default/v1/'+icon2)
        
        df_runetype.append(lst2)
                    
        for j in range(len(df[i]['slots'])) :
            for k in range(len(df[i]['slots'][j]['runes'])) :
                lst = []
                lst.append(df[i]['slots'][j]['runes'][k]['id'])
                lst.append(df[i]['id'])
                lst.append(df[i]['slots'][j]['runes'][k]['name'])
                lst.append(df[i]['slots'][j]['runes'][k]['key'])
                icon = str.lower(df[i]['slots'][j]['runes'][k]['icon'])
                lst.append('https://raw.communitydragon.org/latest/plugins/rcp-be-lol-game-data/global/default/v1/'+icon)
                lst.append(df[i]['slots'][j]['runes'][k]['shortDesc'])
                lst.append(df[i]['slots'][j]['runes'][k]['longDesc'])
                
                df_rune.append(lst)
    
    col = ['runetype_id','runetype_name','runetype_key','runetype_icon']
    df_runetypes  = pd.DataFrame(df_runetype,columns = col)
                    
    col = ['rune_id','runetype_id','rune_name','rune_key','rune_icon','rune_shortdesc','rune_longdesc']
    df_runes  = pd.DataFrame(df_rune,columns = col)

    return df_runes, df_runetypes
                
```

```python
url='http://ddragon.leagueoflegends.com/cdn/13.8.1/data/ko_KR/runesReforged.json'
rune, runetype= runes(url)
```
