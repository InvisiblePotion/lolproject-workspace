```python
url = 'http://ddragon.leagueoflegends.com/cdn/13.8.1/data/ko_KR/summoner.json'
def spell(url):
    df_origin = requests.get(url).json()
    df_keys = list(df_origin['data'].keys())
    
    df_spell = []
    for i in df_keys :
        lst = []
        lst.append(df['data'][i]['key'])
        lst.append(df['data'][i]['id'])
        lst.append(df['data'][i]['name'])
        lst.append(df['data'][i]['description'])
        lst.append('http://ddragon.leagueoflegends.com/cdn/13.8.1/img/spell/'+i+'.png')
        
        df_spell.append(lst)
    
    col = ['spell_id','spell_key','spell_name','spell_desc','spell_icon']
    df_spells = pd.DataFrame(df_spell, columns = col)
    return df_spells
```
